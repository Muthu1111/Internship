1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank=[]
name=[]
artist=[]
views=[]
upload_date=[]
count=0
```

```python
#Scrape the details of most viewed videos on YouTube
name_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in name_tags:
    count+=1
    rank.append(count)
    j=re.split('"',i.text)[1]
    name.append(j)
artist_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in artist_tags:
    artist.append(i.text)
views_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in views_tags:
    views.append(i.text)
upload_date_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in upload_date_tags:
    upload_date.append(i.text)
```

```python
#Check the length of scraped data
print(len(rank),len(name),len(artist),len(views),len(upload_date))
```

```python
#Create Data Frame and display the scraped data
df=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Upload Date':upload_date,'Views(billions)':views})
df
```

4. Scrape the details of trending repositories on Github.com.

Url = https://github.com/

You have to find the following details:  

A) Repository title

B) Repository description

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://github.com/")
```

```python
#Hover over the menu option "Open Source"
open_source=driver.find_element(By.XPATH,'//button[contains(text(),"Open Source")]')
ActionChains(driver).move_to_element(open_source).perform()
#Click the sub-menu option "Trending"
trending=driver.find_element(By.XPATH,'//a[contains(text(),"Trending")]')
trending.click()
```

```python
title=[]
description=[]
contributors=[]
language=[]
```

```python
#Scrape the required details
rows=driver.find_elements(By.XPATH,'//article[@class="Box-row"]')
for row in rows:
    i=row.find_element(By.TAG_NAME,'h2')
    title.append(i.text)
    try:
        i=row.find_element(By.TAG_NAME,'p')
        description.append(i.text)
    except NoSuchElementException:
        description.append('-')
contributors_tags=driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/a[1]')
for i in contributors_tags:
    contributors.append(i.text)
language_tags=driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]')
for i in language_tags:
    try:
        j=i.find_element(By.XPATH,'.//span/span[@itemprop="programmingLanguage"]')
        language.append(j.text)
    except NoSuchElementException:
        language.append('-')
```

```python
#create Data Frame and display the scraped data
df=pd.DataFrame({'Repository Title':title,'Repository Description':description,'Contributors Count':contributors,'Language Used':language})
df
```

8. Details of Datasets from UCI machine learning repositories.

Url = https://archive.ics.uci.edu/ You have to find the following details:

A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute

G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
#Click on the menu option "Datasets"
datasets=driver.find_element(By.XPATH,'//ul[@class="menu menu-horizontal"]/li[1]')
datasets.click()
```

```python
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
no_of_instances=[]
no_of_attributes=[]
year=[]
```

```python
#Click on "EXPAND ALL"
expand_all=driver.find_element(By.XPATH,'//span[@class="swap-on text-primary-content"]')
expand_all.click()
#Scrape the required data
while True:
    dataset_name_tags=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in dataset_name_tags:
        dataset_name.append(i.text)
    rows=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]')
    for row in rows:
        i=row.find_elements(By.TAG_NAME,'span')
        task.append(i[0].text)
        data_type.append(i[1].text)
        no_of_instances.append(i[2].text)
        no_of_attributes.append(i[3].text)
    rows=driver.find_elements(By.XPATH,'//tbody[@class="border"]')
    for row in rows:
        i=row.find_elements(By.TAG_NAME,'td')
        attribute_type.append(i[1].text)
        year.append(i[2].text)
    next=driver.find_element(By.XPATH,'//button[@aria-label="Next Page"]')
    if next.is_enabled():
        next.click()
        time.sleep(4)
        #Click on "COLLAPSE ALL"
        collapse_all=driver.find_element(By.XPATH,'//span[@class="swap-off text-primary-content"]')
        collapse_all.click()
        #Click on "EXPAND ALL"
        expand_all=driver.find_element(By.XPATH,'//span[@class="swap-on text-primary-content"]')
        expand_all.click()
    else:
        break
```

```python
#Check the length of scraped data
print(len(dataset_name),len(data_type),len(task),len(attribute_type),len(no_of_instances),len(no_of_attributes),len(year))
```

```python
#Create Data Frame and display the scraped data
df=pd.DataFrame({'Dataset Name':dataset_name,
                 'Data Type':data_type,
                 'Task':task,
                 'Attribute Type':attribute_type,
                 'No of Instances':no_of_instances,
                 'No of Attributes':no_of_attributes,
                 'Year':year
                })
df
```
