1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year of release) and make data frame.

```python
def display_top_100_movies(url):
    page=requests.get(url)
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        movies=soup.find_all('div',class_="lister-item-content")
        names=[]
        ratings=[]
        years=[]
        for movie in movies:
            names.append(movie.find('a').text.strip())
            rating=movie.find('span',class_="ipl-rating-star__rating").text.strip()
            ratings.append(float(rating))
            year=movie.find('span',class_="lister-item-year").text.strip()
            years.append(re.sub('\D','',year))
        df=pd.DataFrame({'Name':names,'Rating':ratings,'Year of Release':years})
        print(df)
    else:
        print("Can't fetch the data.")
url='https://www.imdb.com/list/ls056092300/'
display_top_100_movies(url)
```

6) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/bestseller?sort=popular .

```python
def product_details(url):
    page=requests.get(url)
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        product_name=[]
        i=1
        for j in soup.find_all('h2',class_="clr-shade4"):
            if i==11:
                break
            else:
                product_name.append(j.text)
                i+=1
        price=[]
        i=1
        for j in soup.find_all('div',class_="discountedPriceText"):
            if i==11:
                break
            else:
                price.append(j.text)
                i+=1
        image_url=[]
        i=1
        for j in soup.find_all('div',class_="productImg"):
            if i==11:
                break
            else:
                image_url.append(j.find('img')['src'])
                i+=1
        for i in range(0,10):
            print("Product Name: ",product_name[i])
            print("Price: ",price[i])
            print("Image URL: ",image_url[i],"\n")
    else:
        print("Cant fetch the data.")

product_details('https://www.bewakoof.com/bestseller?sort=popular%20.')
```

7) Please visit https://www.cnbc.com/world/?region=world and scrap-

a) headings 

b) date 

c) News link

```python
page=requests.get('https://www.cnbc.com/world/?region=world')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
heading=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    heading.append(i.text)
date=[]
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    date.append(i.text)
news_link=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    news_link.append(i['href'])
df=pd.DataFrame({'Heading':heading,'Date':date,'News Link':news_link})
for index,row in df.iterrows():
    print(f"Heading: {row['Heading']}")
    print(f"Date: {row['Date']}")
    print(f"News Link: {row['News Link']}\n")
```

8) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded- articles/ and scrap-

a) Paper title

b) date

c) Author

```python
page=requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
paper_title=[]
for i in soup.find_all('h2',class_="h5 article-title"):
    paper_title.append(i.text.strip())
date=[]
for i in soup.find_all('p',class_="article-date"):
    date.append(i.text)
author=[]
for i in soup.find_all('p',class_="article-authors"):
    author.append(i.text)
df=pd.DataFrame({'Paper Title':paper_title,'Date':date,'Author':author})
for index,row in df.iterrows():
    print(f"Paper Title: {row['Paper Title']}")
    print(f"Date: {row['Date']}")
    print(f"Author: {row['Author']}\n")
```
